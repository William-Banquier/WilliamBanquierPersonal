```python
#importing packages

import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#reading file

train = pd.read_csv('AppleStore.csv')
```

```python
train.head()
```

```python
train.drop("track_name",axis=1,inplace=True)
train.drop("id",axis=1,inplace=True)
train.drop("size_bytes",axis=1,inplace=True)
train.drop("user_rating_ver",axis=1,inplace=True)
train.drop("ver",axis=1,inplace=True)
train.drop("ipadSc_urls.num",axis=1,inplace=True)
train.drop("sup_devices.num",axis=1,inplace=True)
train.drop("vpp_lic",axis=1,inplace=True)
train.drop("rating_count_ver",axis=1,inplace=True)
train.drop("num",axis=1,inplace=True)
train.drop("currency",axis=1,inplace=True)
```

```python
train.head()
```

```python
train.groupby('prime_genre')['user_rating'].count().sort_values(ascending=False)
```

```python
#Making names to numbers

def fillGenre(prime_genre):
    prime_genre = prime_genre[0]
    if prime_genre=='Games':
        return 0
    elif prime_genre=='Entertainment':
        return 1
    elif prime_genre=='Education':
        return 2
    elif prime_genre=='Photo & Video':
        return 3
    elif prime_genre=='Utilities':
        return 4
    elif prime_genre=='Health & Fitness':
        return 5
    elif prime_genre=='Productivity':
        return 6
    elif prime_genre=='Social Networking':
        return 7
    else:
        return 8
train['prime_genre']=train[['prime_genre']].apply(fillGenre,axis=1)
```

```python
train.head()
```

```python
train.groupby('cont_rating')['user_rating'].count().sort_values(ascending=False)
```

```python
def fillGenre(cont_rating):
    cont_rating = cont_rating[0]
    if cont_rating=='4+':
        return 0
    elif cont_rating=='12+':
        return 1
    elif cont_rating=='9+':
        return 2
    else:
        return 4
train['cont_rating']=train[['cont_rating']].apply(fillGenre,axis=1)
```

```python
train.head()
```

```python
#changing floats to ints

def printNull(varible):
    varible = varible[0]
    if varible == None:
        print(varible)
    try:
        return int(varible)
    except:
        return 0
```

```python
train['user_rating']=train[['user_rating']].apply(printNull,axis=1)
train['price']=train[['price']].apply(printNull,axis=1)
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
dummieStars = pd.get_dummies(train['user_rating'],drop_first=False)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train.drop('user_rating',axis=1),
                                                    train['user_rating'],test_size=0.3,
                                                    random_state=101)
                                                    
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
print(logmodel.score(X_train,y_train))
```

```python
#showing classification_report

from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
