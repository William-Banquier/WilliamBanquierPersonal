```python
import pandas as pd
import tkinter as tk
```

```python
df = pd.read_csv('babynames-clean.csv')
```

```python
names = (df['Name'])
```

```python
for i in range(0,len(names)):
    print(names[i])
```

```python
import random as r
x = r.randrange(0,len(names))
print(names[x])
```

```python
import tkinter as tk

MyWindow = tk.Tk()
MyWindow.geometry("500x550")


button = tk.Button(MyWindow,text=names[x],width=17,height=10,font=('Helvetica', '20'))
button.grid(row=1, column=1)

MyWindow.mainloop()
```
